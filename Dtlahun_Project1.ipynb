```python
#project 1
# Date 07/26/2020
#programer: Dereje Tlahun
# I'm the author of this program


import sys
import math
def cookout():
    hotdogINpackage = 8 # $4 per package 
    burgerInpackage = 8 # $4 per package
    beef_pattyPackage = 6 # $6per package
    bunsInpackage = 12 # $4 per package
    name = input("Waht is the customers name:  \n")
    user = (input("Do you went hotdog, beef_patty or burger yes/no: \n"))
    while True:
        try:
            numPeople = int(input("Please enter the number of people attanding: \n"))
            break
        except ValueError:
            print("Oops!  That was no valid number.  Try again...")
    if user == 'yes':
        #Assume evrybody in the partty takes the same amount of hotdog, burger, or beef patty
        while True:
            try:
                numHotdog = int(input("Please enter the number of hot dogs each people need \n"))
                break
            except ValueError:
                print("Oops!  That is not valid number.  Try again...")
        while True:
            try:
                numBurger = int(input("Pleas enter the number of burger each customer need: \n"))
                break
            except ValueError:
                print("Oops!  That is not valid number.  Try again...")
        while True:
            try:
                numbeef_patty = int(input("Please enter the number of beef patty you need: \n"))
                break
            except ValueError:
                print("Oops!  That was no valid number.  Try again...")
        while True:
            try:
                #asume the customer knows how long it takes to finish
                payTime = int(input("How long the partty will take: \n"))
                break
            except ValueError:
                print("Oops!  That was no valid number.  Try again...")
        while True:
            try:
                #asume the customer knows how long it takes to finish
                discount = float(input("if you have Coupon code enter the price else enter 0: \n"))
                break
            except ValueError:
                print("Oops!  That was no valid number.  Try again...")
            
        # if the party is continuing some body has to order something
        while numHotdog <= 0 and numBurger <=0 and numbeef_patty <= 0:
           
            numBurger = int(input("Pleas enter valid number of burger each customer need: \n"))
            numHotdog = int(input("Please enter valid number of hot: " +\
                      "dogs each people need \n"))
            numbeef_patty = int(input("Please enter valid number of beef patty you need: \n"))
            
               
                    
            #Total single hotdog, burger, bun, andf beef patty needed
        hotdoge_Need = numPeople * numHotdog 
        burger_Need = numPeople * numBurger
        beef_pattyNeed = numPeople * numbeef_patty
        buns_need = hotdoge_Need + burger_Need # both hotdog and burger use the same number of bun. 
        
        #Total hotdog, burger, bun, and beef patty package needed 
        purchased_hotdog_package = math.ceil(float(hotdoge_Need / hotdogINpackage)) 
        purchased_burger_package = math.ceil(float(burger_Need / burgerInpackage))
        purchased_bun_package = math.ceil(float(buns_need / bunsInpackage))
        purchased_BeefPatty_package = math.ceil(float(beef_pattyNeed / beef_pattyPackage ))
        
        #the menimum minimum amount of leftovers
        leftover_Hotdogs = (hotdogINpackage * purchased_hotdog_package) - hotdoge_Need
        leftover_Burgers = (burgerInpackage * purchased_burger_package) - burger_Need
        leftover_Buns = (bunsInpackage * purchased_bun_package) - buns_need
        leftover_BeefPatty = (beef_pattyPackage * purchased_BeefPatty_package) - beef_pattyNeed
        
        # calculate how much the cooker spent for the food 
        cashOut_hotdog = purchased_hotdog_package * 4
        cashOut_Burger = purchased_burger_package * 4
        cashOut_Bun = purchased_bun_package * 4
        cashOut_BeefPatty = purchased_BeefPatty_package *6
         
        #Total “out of pocket” cost of cookout 
        total_food_Cost = cashOut_hotdog + cashOut_Burger + cashOut_BeefPatty + cashOut_Bun
        cookout_charge = float((payTime * 25) + (total_food_Cost * 0.2))
        
        
       
        print("\n")
        print("Hello "+ name + " wellcome to cookout service, here is your calculated cookout cost")
        print("number of people: \t\t" + str(numPeople))
        print("Out of pocket cost:\t\t"+ "${:,.2f}".format(total_food_Cost) )
        print("cookout_charge:\t\t\t" + "${:,.2f}".format(cookout_charge))
       
        print("Subtotal:\t\t\t" + "${:,.2f}".format(cookout_charge  + total_food_Cost))
        print("Discount: \t\t\t" + "${:,.2f}".format(discount) )
        print("Total cookout cost\t\t" + "${:,.2f}".format(total_food_Cost + cookout_charge - discount) )
       
        print("\n")
    
        print("For more detail.............")
        print("\t\t HotDog \t\t Burger \t\t Beef patty\t\t Bun             ")
        print("***********************************************************************************************")
        print("single\t\t" +str(hotdoge_Need)+"\t\t\t "+
              str(burger_Need)+"\t\t\t " + str(beef_pattyNeed)+"\t\t\t "+str(buns_need))
        print("package\t\t"+ str(purchased_hotdog_package) +"\t\t\t " +str(purchased_burger_package) +
              "\t\t\t "+ str(purchased_BeefPatty_package) + "\t\t\t "+ str(purchased_bun_package))
        print("LeftOver\t"+ str(leftover_Hotdogs)+ "\t\t\t "+ str(leftover_Burgers)+ "\t\t\t "+ 
              str(leftover_BeefPatty)+ "\t\t\t "+ str(leftover_Buns))
        print("food Price \t"+"${:,.2f}".format(cashOut_hotdog) + "\t\t\t"+ "${:,.2f}".format(cashOut_Burger)+
              "\t\t\t" + "${:,.2f}".format(cashOut_BeefPatty)+ "\t\t\t"+ "${:,.2f}".format(cashOut_Bun))
        
    else:
        print("There is no cookout here")
        sys.exit()
cookout()
```
